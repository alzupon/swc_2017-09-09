```python
for line in text:
    line = line.strip()
```

```python
text_list = text.split('\n')
```

```python
text_list
```

```python
text_list2 = ' '.join(text_list)
text_list2
```

```python
text_list3 = text_list2.split(' ')
text_list3
```

```python
total_a = 0
for word in text_list3:
    a_counter = 0
    for x in word:
        if x == 'a' or x == 'A':
            a_counter += 1
    print("Number of A's in", word, ":", a_counter)
    total_a = total_a + a_counter
print("Total number of A's:", total_a)
```

```python
text_list4 = ''.join(text_list3)
text_list4 = text_list4.lower()
#text_list4 = sorted(text_list4)
text_list4
```

```python
for char in text_list4:
    for char2 in text_list4:
        if char == char2:
            text_list4.replace(char2,'')
print(text_list4)
```

```python
for char in text_list4:
    text_list4.count(char)
    print('Count of', char, ':', text_list4.count(char))
```
